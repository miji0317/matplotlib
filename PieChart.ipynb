#### Importing Libraries and Data

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 영화 장르 데이터 읽어오기
t_mov= pd.read_csv('datasets/types_movies.csv')
```

```python
t_mov
```

### 파이 차트 Pie Chart
데이터 분포를 백분율 비중으로 가시화

```python
plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'])

plt.axis('equal') # 완벽한 원 형태
# plt.axis('auto')    # 찌그러진 타원 형태

plt.show()
```

### 파이 차트 구성 요소
* <b>wedges(웨지):</b> 파이의 개별 조각을 의미하며, Patch 객체에 해당함
* <b>texts:</b> 레이블을 표현하는 텍스트 리스트
* <b>autotexts:</b>  수치형 비중 값

```python
wedges, texts, autotexts = plt.pie(t_mov['Percentage'], 
                                    labels=t_mov['Sector'],
                                    autopct='%.2f')

plt.axis('equal')


print('Wedges: ', wedges)
print('Texts: ', texts)
print('Autotexts: ', autotexts)
```

#### 파이 차트 맞춤화
색상 및 autopct 속성에 대한 출력 형식을 설정

```python
colors = ['darkorange', 'sandybrown', 'darksalmon', 'orangered','chocolate']

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f')

plt.axis('equal')

plt.show()
```

#### explode(터짐) 속성
- 파이 차트의 특정 웨지를 강조하고 싶다면, explode 옵션을 활용
- explode 값은 각 조각에 대해서, 반지름을 기준으로 떨어지는 간격 비율을 의미함

```python
# explode = (0.1, 0, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Sci-Fi')
# explode = (0, 0.1, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Drama')
explode = (0, 0.1, 0, 0, 1)  # explode the 2nd and the 5th slice

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
```
