#### Importing Libraries and Data

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 주가 데이터 읽어오기
stock_data = pd.read_csv('datasets/stocks.csv')
```

```python
stock_data.head()
```

```python
# stock_data['Date']를 날짜형으로 변환하여 원본에 다시 저장
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.head()
```

#### 주가를 작은 별도 차트로 중복하여 그려서 나란히 비교하는 시각화

```python
fig = plt.figure(figsize=(10,6))

# 녹색으로 AAPL 주가를 표시할 축 정의 [x, y, width, length]
ax1 = fig.add_axes([0, 0, 1, 1])
# 파랑으로 IBM 주가를 표시할 축 정의 [x, y, width, length]
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])


ax1.set_title('AAPL(green) vs IBM(blue, inset)')

ax1.plot(stock_data['Date'],
         stock_data['AAPL'],
         color='green')

ax2.plot(stock_data['Date'],
         stock_data['IBM'], 
         color='blue')
plt.show()

# 이렇게 두 차트를 나란히 비교할 때에는 눈금을 통일해야 함
```

#### (축 눈금 통일하지 않고) 다수 주가를 비교

```python
# (폭, 높이)로 Figsize 지정
fig = plt.figure(figsize=(15,7))

fig.suptitle('Stock price comparison 2007-2017', 
             fontsize=20)

# subplot(rcn) 형식으로 2행 x 3열,  총 6개의 subplot을 구성
ax1 = fig.add_subplot(231)
ax1.set_title('MSFT')

ax1.plot(stock_data['Date'],
         stock_data['MSFT'], 
         color='green')

ax2 = fig.add_subplot(232)
ax2.set_title('GOOG')

ax2.plot(stock_data['Date'],
         stock_data['GOOG'], 
         color='purple')

ax3 = fig.add_subplot(233)
ax3.set_title('SBUX')

ax3.plot(stock_data['Date'],
         stock_data['SBUX'], 
         color='magenta')

ax4 = fig.add_subplot(234)
ax4.set_title('ADBE')

ax4.plot(stock_data['Date'],
         stock_data['ADBE'], 
         color='orange')

ax5 = fig.add_subplot(235)
ax5.set_title('NFLX')

ax5.plot(stock_data['Date'],
         stock_data['NFLX'], 
         color='chocolate')

ax6 = fig.add_subplot(236)
ax6.set_title('ORCL')

ax6.plot(stock_data['Date'],
         stock_data['ORCL'], 
         color='teal')

plt.show()
```

#### (축 눈금 통일하고) 다수 주가를 비교

```python
# (폭, 높이)로 Figsize 지정
def y_axis(ax, ymin, ymax, ydelta):
    yticks = np.arange(ymin, ymax, ydelta)
    ax.set_yticks(yticks)
    yrange = (yticks[0], yticks[-1])
    ax.set_ylim(yrange)


fig = plt.figure(figsize=(15,7))

fig.suptitle('Stock price comparison 2007-2017',
             fontsize=20)

# subplot(rcn) 형식으로 2행 x 3열,  총 6개의 subplot을 구성
ax1 = fig.add_subplot(231)
ax1.set_title('MSFT')
# y_axis(ax1, 0, 120, 20)
y_axis(ax1, 0, 1200, 200)

ax1.plot(stock_data['Date'],
         stock_data['MSFT'],
         color='green')

ax2 = fig.add_subplot(232)
ax2.set_title('GOOG')
y_axis(ax2, 0, 1200, 200)

ax2.plot(stock_data['Date'],
         stock_data['GOOG'],
         color='purple')

ax3 = fig.add_subplot(233)
ax3.set_title('SBUX')
y_axis(ax3, 0, 1200, 200)

ax3.plot(stock_data['Date'],
         stock_data['SBUX'],
         color='magenta')

ax4 = fig.add_subplot(234)
ax4.set_title('ADBE')
y_axis(ax4, 0, 1200, 200)

ax4.plot(stock_data['Date'],
         stock_data['ADBE'],
         color='orange')

ax5 = fig.add_subplot(235)
ax5.set_title('NFLX')
y_axis(ax5, 0, 1200, 200)

ax5.plot(stock_data['Date'],
         stock_data['NFLX'],
         color='chocolate')

ax6 = fig.add_subplot(236)
ax6.set_title('ORCL')
y_axis(ax6, 0, 1200, 200)

ax6.plot(stock_data['Date'],
         stock_data['ORCL'],
         color='teal')

plt.show()
```
